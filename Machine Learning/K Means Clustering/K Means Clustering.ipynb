# K Means Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=300, n_features=2, centers=4, cluster_std=0.6,
                 random_state=0)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,6))
axes[0].set_title('KMeans')
axes[0].scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,
               cmap='viridis')

axes[1].set_title('Original')
axes[1].scatter(data[0][:,0],data[0][:,1],c=data[1],
               cmap='viridis')
```

```python
from sklearn.metrics import pairwise_distances_argmin
import numpy as np
```

```python
def find_clusters(n_clusters):
    rng = np.random.RandomState(seed=2)
    i = rng.permutation(len(data[0]))[:n_clusters]
    centers = data[0][i]
    
    while True:
        labels = pairwise_distances_argmin(data[0], centers)
        new_centers = np.array([data[0][labels==i].mean(0) for i in range(n_clusters)])
        
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels
        
centers, labels = find_clusters(4)

plt.scatter(data[0][:,0], data[0][:,1], c=labels, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.7)
```
