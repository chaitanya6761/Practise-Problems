## Pandas Data Visualizations

```python
#Requried Data And Imports
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline


df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
#Histogram
df1['A'].hist(bins=30)
```

```python
df1['A'].plot.hist()
```

```python
#Area Plots
df2.plot.area()
```

```python
#Bar Plot
df2.plot.bar()
```

```python
#Bar Plots With Stacked Bars
df2.plot.bar(stacked=True)
```

```python
#Line Plot
df1.plot.line(x=df1.index, y='B', figsize=(15,3))
```

```python
#Scatter Plots
df1.plot.scatter(x='A', y='B')
```

```python
df1.plot.scatter(x='A', y='B', c='C')
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100)
```

```python
#Box Plots
df2.plot.box()
```

```python
#Hex Bin Plots
df = pd.DataFrame(np.random.randn(1000,2), columns=['A','B'])
df.head()
```

```python
df.plot.hexbin(x='A', y='B', gridsize=25, cmap='coolwarm')
```

```python
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```

```python
df2.plot.density()
```
