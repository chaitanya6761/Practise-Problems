## Seaborn Regression Plots

```python
#Required Data And Imports
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', markers=['o','v'])
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip',col='sex') 
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip',col='sex', row='time') 
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip',col='day', hue='sex', aspect=0.7) 
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip',col='day', hue='sex', size=8) 
```
