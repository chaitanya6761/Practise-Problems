## Seaborn Dist Plots

```python
#Required Imports
import seaborn as sns
%matplotlib inline

import matplotlib.pyplot as plt
plt.rcParams["patch.force_edgecolor"] = True
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Dist Plot Allows Us To Show The Distribution Of Univariate Set Of Observations
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
#Joint Plot Allows Us To Match Up Distributions Of Bi Variate Data
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
#Pair Polt Creates Joint Plot Of Every Possible Combinations From Numerical Column
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```
