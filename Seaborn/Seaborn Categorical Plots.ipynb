## Seaborn Categorical Plots

```python
#Required Imports And Data
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',estimator=np.std)
```

```python
sns.countplot(tips['sex'])
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split='true')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True, hue='sex')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True, hue='sex', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```
