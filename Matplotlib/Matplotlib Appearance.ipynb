```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red', linewidth=5, alpha=0.5)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red', linewidth=5, linestyle='-.')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.scatter(x,y,color='red')
axes.plot(x,y,color='red')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',lw=1,marker='o',markersize=10,markerfacecolor='yellow')
```
